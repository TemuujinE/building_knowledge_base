```python
import json
import pandas as pd
import numpy as np

import pickle
```

```python
file_path = 'knowledge_base/'

with open('knowledge_base/product_names.pickle', 'rb') as handle:
    product_names = pickle.load(handle)
```

```python
product_names
```

```python
with open((file_path + 'depo.json').format(1), 'r', encoding = 'utf-8') as json_file:
    depo = json.load(json_file)
    
with open((file_path + 'account.json').format(1), 'r', encoding = 'utf-8') as json_file:
    account = json.load(json_file)
    
with open((file_path + 'loan.json').format(1), 'r', encoding = 'utf-8') as json_file:
    loan = json.load(json_file)
    
with open((file_path + 'card.json').format(1), 'r', encoding = 'utf-8') as json_file:
    card = json.load(json_file)
```

```python
def return_info(linkie, product):
    # Get product info
    product_name, about_product, temp_1, temp_2, required_materials, dfs = product_info(linkie, product)
    
    # Print some basic info such as product name and about product
    print('Number of tables on this product:', len(dfs), '\n')
    print(product_name + ': ' + about_product)

    for temp in [temp_1, temp_2]:
        if len(temp):
            print(temp)
            
    # Display tables with product info
    if len(dfs):
        print('\n', 'Tables on this product:')
        
        for df in dfs:
            display(df)
```

```python
def return_product_info(knowledge_base, product, product_name):
    if product == 'depo':
        values = product_names[0]
    elif product == 'account':
        values = product_names[1]
    elif product == 'card':
        values = product_names[2]
    elif product == 'loan':
        values = product_names[3]

    idx = values.index(product_name.lower())
    
    product_name = knowledge_base[idx][0]['product_name']
    about_product = knowledge_base[idx][0]['about_product']
    required_materials = knowledge_base[idx][0]['required_materials']
    additional_info = knowledge_base[idx][0]['additional_info']
    
    dfs = []
    for table in knowledge_base[0][1:]:
        dfs.append(pd.DataFrame(table))
        
    return product_name, about_product, required_materials, additional_info, dfs
```

```python
product_name, about_product, required_materials, additional_info, dfs = return_product_info(depo, 'depo', 'уян хадгаламж')

# Print some basic info such as product name and about product
print('Number of tables on this product:', len(dfs), '\n')
print(product_name + ': ' + about_product)

for info in additional_info:
    if len(info):
        print(info)

# Display tables with product info
if len(dfs):
    print('\n', 'Tables on this product:')

    for df in dfs:
        display(df)
```
