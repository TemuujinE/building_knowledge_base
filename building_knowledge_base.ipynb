```python
def return_info(linkie, product):
    # Get product info
    product_name, about_product, temp_1, temp_2, required_materials, dfs = product_info(linkie, product)
    
    # Print some basic info such as product name and about product
    print('Number of tables on this product:', len(dfs), '\n')
    print(product_name + ': ' + about_product)

    for temp in [temp_1, temp_2]:
        if len(temp):
            print(temp)
            
    # Display tables with product info
    if len(dfs):
        print('\n', 'Tables on this product:')
        
        for df in dfs:
            display(df)
```

# Building knowledge base

```python
import json

products = [depo[0], account[0], card[0], loan[0]]
product_links = [depo[2], account[2], card[2], loan[2]]

for i in range(len(products)):
    knowledge_base = []
    product = products[i]
    linkie = product_links[i]
    
    for link in linkie:
        product_name, about_product, temp_1, temp_2, required_materials, dfs = product_info(link, product)

        for idx in range(len(dfs)):
            dfs[idx] = dfs[idx].to_dict()

        dfs.insert(0, {'product_name': product_name, 
                       'about_product': about_product, 
                       'required_materials': required_materials,
                       'additional_info': [temp_1, temp_2]})    
        knowledge_base.append(dfs)
        
        
        
        file_save_name = 'knowledge_base/' + product + '.json'
        
    with open(file_save_name.format(1), 'w', encoding = 'utf-8') as file:
        json.dump(knowledge_base, file, ensure_ascii = False, indent = 4)
        
    print(f'Number of {product} products:', len(knowledge_base))
```
